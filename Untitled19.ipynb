```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

import pandas as pd
from scipy.stats import ttest_ind, ttest_rel 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/s-a-nersisyan/HSE_bioinformatics_2021/master/seminar12/colon_cancer_tumor_vs_normal_paired_FPKM.tsv', sep = '\t', index_col = 0)
df
# –ø–µ—Ä–≤—ã–µ 5 -- –Ω–æ—Ä–º–∞–ª—å–Ω–∞—è —Ç–∫–∞–Ω—å, 5-10 -- –æ–ø—É—Ö–æ–ª–µ–≤–∞—è
```

```python
df['ttest_rel'] = [ttest_rel(df.loc[gene].iloc[0:5], df.loc[gene].iloc[5:10])[1] for gene in df.index]
df['ttest_ind'] = [ttest_ind(df.loc[gene].iloc[0:5], df.loc[gene].iloc[5:10])[1] for gene in df.index]
```
